1. What is Python, and why is it popular ?
 - Python is a programming language which has special set of rules and symbols that we use to give instructions to the computer.
 - It is very popular as widely used i industry, also has a lot of libraries, easy to learn, automation, databases, and for its versatlity and readability.

2. What is an interpreter in Python ?
  - In python, an interpreter means that program reads and execute the code line by line without compiling the code.we can directly run python files with the interpreter.

3. What are pre-defined keywords in Python ?
  - Pre-deefined keywords in python are reserved words that holds a special meaning & have a specific purpose.
  Ex: if, else, true, false, break, continue, etc.

4. Can keywords be used as variable names ?
  - No, keywords are pre-defined or reserved words that have speific meaning and purpose. If we use keywords as variable name then syntax error occur as the interpreter gets confused and not able to distinguish between keywords and variable name.

5. What is mutability in Python ?
  -The ability of Data structures to change their value after they are created s called mutability and these data structures are known as mutable data structures or objects.

6. Why are lists mutable, but tuples are immutable ?
  - in lists, we can do changes like we can add value to a list after creation or remove a value, etc. so they are called mutable type.
  But in tuples, we cannot do any changes so they are called immutable type.

7. What is the difference between “==” and “is” operators in Python?
  -  "==" and "is" operators, both are used for comparison but "==" operator is used to check the equality of values while "is" operator used for identity, it compares the memory address of the object.

8. What are logical operators in Python ?
  - Logical operators are used to combine two or more operands and do logical comparison using boolean expression.
  Three types of logical operator: AND, OR & NOT.

9. What is type casting in Python ?
  - The process of changing the data types of an object or value is known as typecasting.

10. What is the difference between implicit and explicit type casting ?
  - In impicit type cating, python by default cosider a particular data type for a particular value while in explicit type casting, user converts the data type of particular value using in-built functions.

11. What is the purpose of conditional statements in Python ?
  - Conditional statements help us to code the program decision based on some pre-conditions. it compare the value based on conditions and give the result.

12.  How does the elif statement work ?
  -Elif stands for 'else if' and used in python programming to test muliple conditions to get the result.

13. What is the difference between for and while loops ?
  - 'for' loop iterates over a sequence like a range or list, while the 'while' loop repeats based o the conditions applied.

14. Describe a scenario where a while loop is more suitable than a for loop.
  - While loop is best used when we don't know the number of iterations in advance.

```python
# Write a Python program to find the sum of all even numbers between 1 and 50.
sum=0
for i in range(1,51):
  if i%2==0:
    sum=sum+i
print(sum)
```

```python
# Write a program to reverse a string using a while loop.
string=input("Enter a string: ")
reverse=""
i=len(string)-1
while i>=0:
  reverse=reverse+string[i]
  i=i-1
print(reverse)
```

```python
#. Write a Python program to calculate the factorial of a number provided by the user using a while loop
num=int(input("Enter a number: "))
fact=1
i=1
while i<=num:
  fact=fact*i
  i=i+1
print(fact)
```
